```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import StratifiedKFold
```

```python
df=pd.read_csv('Zoo.csv')
```

```python
df
```

# splitting the target and the features

```python
target=df[['type']]
```

```python
target

```

```python
features=df.drop(columns='type')
```

```python
features
```

```python
features.drop(columns='animal name',inplace=True)
```

```python
features.columns
```

```python
features.isnull().sum()
```

```python
features['legs'].unique()
```

# split the train and the testing data

```python
x_train,x_test,y_train,y_test=train_test_split(features,target,train_size=0.8,random_state=50)
```

```python
x_train.shape
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train,y_train)
```

```python
y_predict=knn.predict(x_test)
```

```python
y_predict
```

# accuracy score

```python
accuracy_score(y_test,y_predict)
```

# classification report

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_predict))
```
